```python
# carregando dados das ações 

import yfinance as yf

ticker = input("Digite o código da ação: ")
dados = yf.Ticker(ticker).history("2y")
dados.head

dados["Close"]
```

#### Tratamento de dados

```python
dados.head()
```

```python
# resetando o índice 
treinamento = dados.reset_index()

# isolando as colunas desejadas
treinamento = treinamento[["Date", "Close"]]

# limpando a coluina data, tirando o time zone(tz)
treinamento["Date"].dt.tz_localize(None)
treinamento["Date"] = treinamento["Date"].dt.tz_localize(None)

# substituindo o nome das colunas para o treinamento dos dados
treinamento.columns = ['ds', 'y']
treinamento
```

#### Treinando o Modelo de Machine Learning

```python
# instalando o prophet
!pip install prophet

from prophet import Prophet
from prophet.plot import plot_plotly

# criar o modelo de inteligência artificial - Machine Learning
modelo = Prophet()

# treinar modelo
modelo.fit(treinamento)

```

#### Realizar as previsões

```python
# definindo o periodo futuro da previsão
periodo = modelo.make_future_dataframe(90)

previsoes = modelo.predict(periodo)
```

#### Gerando a visualização gráfica

```python
plot_plotly(modelo, previsoes)
```
